```python
from gensim.models.word2vec import Text8Corpus
from gensim.models import Word2Vec

def get_model(path='data/data.txt', save='data/word_model.txt'):
    sentences = Text8Corpus(path)
    model = Word2Vec(sentences)
    model.wv.save_word2vec_format(save, binary=False)
    print(model.wv)
    return model.wv
wv = get_model()
```

```python
import re
import jieba
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sympy import apart

def pre_process(path='data/data1.txt'):
    labels = []
    texts = []
    corpus = []
    value = re.compile(r'^[\u4e00-\u9fa5]{2,}$')
    try:
        stopwords = open('data/stopwords.txt')
        for word in stopwords:
            word = word.strip()
            texts.append(word)
    except: 
        print('error')
    
    try:
        with open(path) as f:    
            line = f.readline()
            while line:
                info = line.split('_!_')
                label = info[1]
                if label == '102':
                    labels.append(0)
                elif label == '103':
                    labels.append(1)
                elif label == '108':
                    labels.append(2)
                elif label == '109':
                    labels.append(3)
                data1 = jieba.cut(info[3])
                data2 = info[4].split(',')
                data_adj = []
                # for item in data1:
                #     if item not in texts:
                #         if value.match(item):
                #             data_adj.append(item)
                for item in data2:
                    if item not in texts:
                        if value.match(item):
                            data_adj.append(item)
                corpus.append(data_adj)
                line = f.readline()
    except:
        print('error')
    
    return corpus, labels

corpus, labels = pre_process()
```

```python
import numpy as np
import pandas as pd

def get_word2vec_vec(corpus, wv):
    text_vec = np.zeros((len(corpus), 100))
    index = 0
    for text in corpus:
        # print(text)
        vec = np.zeros((1, 100))
        for word in text:
            try:
                vec += wv[word]
            except:
                pass
        if len(text) == 0:
            continue
        text_vec[index] = vec/len(text)
        index += 1
    return text_vec
    
weight = get_word2vec_vec(corpus, wv)
```

```python
from sklearn.cluster import KMeans

class kmeans:
    def __init__(self, weight, labels, clusters=4):
        self.weight = weight
        self.labels = labels
        self.clusters = clusters
        kmeans = KMeans(n_clusters=clusters)
        self.result = kmeans.fit_predict(weight)
    def get_result(self):
        X = [[0 for i in range(self.clusters)] for i in range(self.clusters)]
        for i in range(len(self.result)):
            predict = self.result[i]
            label = labels[i]
            X[predict][label] += 1
        for i in range(self.clusters):
            self.result[self.result==i] = X[i].index(max(X[i]))+10
        self.result -= 10
        # print(X)
    def get_correct(self):
        total = len(self.result)
        count = 0
        for i in range(total):
            predict = self.result[i]
            label = labels[i]
            if predict == label:
                count += 1
        print(count/total)

a = kmeans(weight=weight, labels=labels)
a.get_result()
a.get_correct()
```
