```python
import numpy as np
```

```python
def f(x,n):
    r = 0
    for i in range(n-1):
        x_i,x_i1 = x[i][0],x[i+1][0]
        r += (1-x_i)**2 + 100*(x_i1-x_i**2)**2
    return r
```

```python
def get_g(x,n):
    d = np.zeros((n,1))
    for i in range(n-1):
        x_i,x_i1 = x[i][0],x[i+1][0]
        d[i][0] += 2*(x_i -1) + 400*x_i*(x_i**2-x_i1)
        d[i+1][0] += 200*(x_i1-x_i**2)
    return d
```

```python
def msgd(gamma,eta,n,iter):
    xs,fs,gs = [],[],[]
    x = np.zeros((n,1))
    v = np.zeros((n,1))
    for i in range(iter):
        xs.append(x)
        fs.append(f(x,n))
        g = get_g(x,n)
        gs.append(np.sqrt(np.dot(g.T,g)))
        v = gamma*v - eta*g
        x = x + v
    return x,xs,fs,gs
    
```

```python
x,xs,fs,gs = msgd(0.9,0.001,2,1000)
```

```python
print(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
step = 0.5
x = np.arange(0,2,step)
y = np.arange(0,2,step)
X,Y = np.meshgrid(x,y)
Z = (1-X)**2 + 100*(Y-X**2)**2
```

```python
plt.figure(figsize=(6,6))
plt.contour(X,Y,Z,50)
xx = [a[0][0] for a in xs]
yy = [a[1][0] for a in xs]
l = len(xx)
nums = 100
step = l//nums
idxs = [i for i in range(0,l,step)]
if l%step != 0:idxs.append(l-1)
x_n = [xx[i] for i in idxs]
y_n = [yy[i] for i in idxs]
plt.plot(x_n,y_n,color='r',marker='o',markerfacecolor='blue',markersize=12)
plt.show()
```
