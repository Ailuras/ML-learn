```python
import numpy as np
```

```python
def f(x):
    return 0.1*x[0][0]**2 + 2*x[1][0]**2
```

```python
def get_g(x):
    g = np.zeros((2,1))
    g[0][0] = 0.2*x[0][0]
    g[1][0] = 4*x[1][0]
    return g
```

```python
def nag(gamma,eta,iter):
    x = 5*np.ones((2,1))
    v = np.zeros((2,1))
    xs = []
    for i in range(iter):
        xs.append(x)
        xt = x + gamma*v
        g = get_g(xt)
        v = gamma*v - eta*g
        x = x + v
    return x,xs
```

```python
x,xs0 = nag(0.9,0.01,500)
print(x)
x,xs1 = nag(0.6,0.01,500)
print(x)
x,xs2 = nag(0.9,0.001,2000)
print(x)
x,xs3 = nag(0.6,0.001,2000)
print(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
step = 0.5
x = np.arange(-6,6,step)
y = np.arange(-6,6,step)
X,Y = np.meshgrid(x,y)
Z = 0.1*X**2 + 2*Y**2
plt.figure(figsize=(15,15))
ax = plt.subplot(221)
ax.set_title('epoch=500,gamma = 0.9,eta = 0.01')
plt.contour(X,Y,Z,30)
xx = [a[0][0] for a in xs0]
yy = [a[1][0] for a in xs0]
plt.plot(xx,yy,color='r',marker='o',markerfacecolor='blue',markersize=12)
ax = plt.subplot(222)
ax.set_title('epoch=500,gamma = 0.6,eta = 0.01')
plt.contour(X,Y,Z,30)
xx = [a[0][0] for a in xs1]
yy = [a[1][0] for a in xs1]
plt.plot(xx,yy,color='r',marker='o',markerfacecolor='blue',markersize=12)
ax = plt.subplot(223)
ax.set_title('epoch=2000,gamma = 0.9,eta = 0.001')
plt.contour(X,Y,Z,30)
xx = [a[0][0] for a in xs2]
yy = [a[1][0] for a in xs2]
plt.plot(xx,yy,color='r',marker='o',markerfacecolor='blue',markersize=12)
ax = plt.subplot(224)
ax.set_title('epoch=2000,gamma = 0.6,eta = 0.001')
plt.contour(X,Y,Z,30)
xx = [a[0][0] for a in xs3]
yy = [a[1][0] for a in xs3]
plt.plot(xx,yy,color='r',marker='o',markerfacecolor='blue',markersize=12)
plt.show()
```
